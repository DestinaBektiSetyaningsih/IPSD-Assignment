### Menggunakan Visualisasi

#### Visualisasi Data 1

```python
df = pd.read_csv('/content/train_features_1.csv')
```

```python
# Distribusi Data 1 "Pendidikan"
class_counts = df['pendidikan'].value_counts()
labels = class_counts.index
sizes = class_counts.values
colors = ['lightcoral', 'lightgreen', 'lightskyblue']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Pie Chart Distribusi Kelas Pendidikan')
plt.show()
```

program diatas yaitu untuk mengetahui distribusi kelas dalam pendidikan yang terdapat pada data features 1, dan kita tahu bahwa pendidikan tertinggi ada di pendidikan sarjana dengan memperoleh nilai 53.5%, dan terendah di pendidikan SMP dengan memperoleh nilai 1,7% atau hanya 0.6%.

```python
# Distribusi Data 1 "status_pernikahan"
class_counts = df['status_pernikahan'].value_counts()
labels = class_counts.index
sizes = class_counts.values
colors = ['lightgrey', 'lightpink', 'skyblue']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Pie Chart Distribusi Kelas Status Pernikahan')
plt.show()
```

program diatas yaitu untuk mengetahui distribusi kelas dalam status pernikahan yang terdapat pada data features 1, dan kita tahu bahwa status pernikahan tertinggi ada di rencana menikah ,  dengan memperoleh nilai 35,3%, dan terendah di status pernikahan ada di cerai mati dengan memperoleh nilai 0.0 atau hanya 1,1%.

```python
# Histogram Distribusi Data 1 "Tahun Kelahiran" dan "Pendapatan"
plt.figure(figsize=(10, 6))
plt.hist(df['tahun_kelahiran'].dropna(), bins=30, color= 'black', edgecolor='pink')

plt.title('Histogram Distribusi Tahun Kelahiran dan Pendapatan ')
plt.xlabel('tahun_kelahiran')
plt.ylabel('pendapatan')

plt.grid(axis='y', alpha=0.75)
plt.show()
```

program diatas yaitu distribusi yang berbentuk histogram yang dimana untuk mengetahui distribusi kelas dalam tahun kelahiran serta pendapatannya yang terdapat pada data features 1, dan kita tahu bahwa pendapatan tertinggi ada di tahun 1970 an, dengan memperoleh pendapatan sebanyak 490 dan terendah di pendapatan ada di 0 dengan memperoleh tahun kelahiran 1910an dan 1920, dan dengan rentang 0,1 di tahun 1900 dan 2000.

```python
# Scatter Plot Distribusi Data 1 "Tahun Kelahiran" dan "Pembelian Diskon"
df = pd.read_csv('/content/train_features_1.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['terakhir_belanja'], df['pembelian_diskon'], alpha=0.5, color='coral', edgecolor='black')
plt.title('Scatter Plot Terakhir Belanja vs Pembelian Diskon')
plt.xlabel('terakhir belanja')
plt.ylabel('pembelian_diskon')

plt.grid(alpha=0.75)
plt.show()
```

Program diatas yaitu distribusi yang berbentuk scatter plot yang dimana untuk mengetahui distribusi kelas dalam tahun terakhir belanja serta pembelian dengan diskon yang terdapat pada data features 1, dan kita tahu bahwa semakin banyak titik-titik yang berkumpul menandakan bahwa nilai korelasi dari data tersebut kuat. Semakin rapat titik-titik data berkumpul, maka semakin kuat nilai korelasinya.

Kesimpulan dari scatter plot diatas yaitu nilai korelasi terbanyak/terkuat ada di pembelian dengan diskon 0.0 dan terakhir belanja ada di rentang 100 an lebih, sedangakn untuk korelasi terdikit/terlemah ada di pembelian dengan diskon 20.0 dan terakhir belanja ada direntang 70 an.
