### 1. Imputasi missing value dengan mean, median, dan modus

```python
import pandas as pd
```

```python
# Load data
df = pd.read_csv('/content/diabetes.csv')
print(df)
```

```python
# Menampilkan data SEBELUM melakukan imputasi
print("Data sebelum melakukan imputasi:")
print(df)
```

```python
# Memeriksa data yang missing value
df.isnull().any()
```

```python
# Menghitung data yang missing value
df.isnull().sum()
```

```python
# Menampilkan informasi dataframe
df.info()
```

##### Mengimputasi missing value dengan data yang bernilai NaN menggunakan mean, median, dan modus

```python
df.head()
```

```python
# Imputasi missing value dengan menggunakan mean
for column in df.select_dtypes(include=['number']):
    df[column] = df[column].fillna(df[column].mean())
```

```python
# Imputasi missing value dengan menggunakan median \
for column in df.select_dtypes(include=['number']):
    df[column] = df[column].fillna(df[column].median())
```

```python
# Imputasi missing value dengan menggunakan modus
for column in df.select_dtypes(include=['number']):
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
# Menampilkan data SESUDAH melakukan imputasi
print("\nData sesudah melakukan imputasi:")
print(df)
```

### 3. Lakukan imbalance handling dengan undersampling

```python
#pip install Imbalanced-Learn
```

```python
# Import librabry
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.under_sampling import RandomUnderSampler, TomekLinks
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
# Load data
data = load_diabetes()
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
# Mengubah label numerik ke kategori yang sesuai
df['Outcome'] = df['Outcome'].map({0: 'No Diabetes', 1:'Diabetes'})
```

```python
# Menampilkan jumlah setiap kelas pada kolom target
df['Outcome'].value_counts()
```

```python
# Mencari data x= fitur dan y= target
x = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
# Melakukan random Under Sampling
runs = RandomUnderSampler(random_state=42)
x_resampled, y_resampled = runs.fit_resample(x, y)
```

```python
# Memeriksa distribusi kelas pada kolom target SEBELUM melakukan undersampling
y.value_counts()
```

```python
# Memeriksa distribusi kelas pada kolom target SESUDAH melakukan undersampling
y_resampled.value_counts()
```
